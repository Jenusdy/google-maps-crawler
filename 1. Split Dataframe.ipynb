### Split Dataframe

Untuk memudahkan pembagian tugas, maka dataframe harus displit terlebih dahulu

Running ini hanya sekali saja, biar nggak nimpa dataframe yang sudah diproses

```python
# Ganti Path ke folder input
df = pd.read_csv('input/20260313 - SBR Jatinegara.csv', sep=';', dtype=object)

# Ganti Row sesuai dengan Pembagian
row_start = 1
row_end = 3
```

```python
df = df.iloc[row_start -1 : row_end]
```

```python
df.to_csv(f'proses/Data-SBR-{row_start}-{row_end}.csv', index=False, sep=';')
```
